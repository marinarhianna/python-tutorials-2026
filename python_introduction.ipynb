As you can see, it is tedious and clunky to write out that command for every planet in the list.

A **loop** is a nice way to make this quicker and more elegant.

```python
# a loop for printing out how many letters are in the name of every thing in our list
for thing in planets:
  print("The number of letters in", thing, "is", len(thing))
```

Recipe for writing a loop 🤌😋


*   Begin the loop with `for`
*   The word `thing` in this loop can actually be any word at all, as long as it matches what is written inside the print statement within the loop.
* Tell Python which list to go through by saying `in` followed by the name of the list.
* Finish with a colon `:`
* Ensure the command you are looping over is indented. The indent should appear automatically when you press enter after the colon.

Loops can be made more complex.

Below we have a new list, containing both the planet name and the radius of the planet.

Notice how commas and different brackets are used to store this list.

```python
# planetary radii (in km)
planet_info = [("Mercury", 2440),
    ("Venus", 6052),
    ("Earth", 6378),
    ("Mars", 3390),
    ("Jupiter", 69911),
    ("Saturn", 58232),
    ("Uranus", 25362),
    ("Neptune", 24622)]
```

We can specify more than one thing listed for the loop to go through.

```python
# loop through both the name and the radius
for name, radius in planet_info:
    print("The planet", name, "has a radius of", radius, "km")
```

⛳ **See for yourself:** Use the cell below to write a loop exactly like the one above, but replace `name` and `radius` with whatever words you want.

Let's think more closely about why indexing can be helpful.

The cell below uses indexes to define Earth's radius.

```python
earth_radius = planet_info[2][1]
```

**Discussion:**
* Why does `[2][1]` indexed onto `planet_info` give us the radius of Earth?
* What index would we use if we wanted (a) Saturn's radius, (b) Jupiter's name.

We can also use loops for other tasks, not only printing.

```python
for planet, radius in planet_info:
  new_radius = radius + earth_radius

  print("The radius of Earth plus the radius of", planet, "is", new_radius)
```

⛳ **CHALLENGE**: Write a loop that prints out the ratio of each planet's radius to Earth's radius.

```python
#@title Challenge Solution:

for name, radius in planet_info:
  radius_ratio = radius / earth_radius

  print("The radius of", name, "is", radius_ratio, "times the radius of Earth")
```

### **Exercise 4** - Functions 🖥

A function is a piece of code that we can re-use many times without having to rewrite all the commands.

Recipe for defining a function 🤌😋

* Begin the command with `def`
* Name the function and give it an input.
* Finish with a colon `:`
* Indent after the `:` and write the code you want to be able to repeat here.

The purpose of the function below is to print out a message such as "Hello, Marina". Read the code, try to see how it matches the recipe, and run it.

```python
def greet(person):            # defining the function, naming it "greet", and making the input a person
  print("Hello,", person)     # this is the line of code we want to be able to re-use without writing it out many times
```

You will see that the code ran, but nothing printed out. This is because all we did above was *define* the function. We haven't actually *used* it.

To use a function, all we need to know is the name and what the input is.

* Name of function is `greet`.
* The input is a person, so we can use Marina as an example.


```python
greet("Marina")

# have a go at changing the person inside the function and re-running the cell!
```

Functions can be used to do many things in one go, and are fully customisable to do whatever we want them to.

Another very simple example:
* Name of function is `add`
* Inputs are two numbers
* The code we want to easily re-use is adding the two numbers together.

```python
# example function for adding two numbers together
def add(num1, num2):
  return num1 + num2
```

```python
add(2, 5)    # try changing these numbers
```

We can also put a loop inside of a function.

The function below is for counting down from a number:
* Name is `countdown`
* Input is `start`, ie. where do we start counting down from.
* The piece of code we want to re-use is our loop, and then a print statement.

This loop:
* Instead of a pre-defined list (like our planet list before), we can just define a range of numbers that start from our input `start`, end at `0`, and move in steps of `-1` (like how we would count down).

```python
def countdown(start):
  for number in range(start, 0, -1):
    print(number)
  print("Lift off!")
```

Below is a different function. Read the code and comments carefully before running it.

```python
def sum_up(list_of_numbers):
  total = 0          # set total to zero initially, which we will then add to

  for num in list_of_numbers:
    total += num               # += just means "add to the existing total"
  return total
```

```python
#@title Challenge Solution (Example):

list_of_numbers = [1, 2, 3, 4, 5]
sum_up(list_of_numbers)
```

### **Exercise 5** - Libraries 📖


* A Python library is a ready-made collection of pre-defined functions that we can import and use.

* It makes coding much quicker than having to define every function ourselves!

* Of course, there are still times where we do need to define our own unique ones, but libraries help us out!

We can give our imported libraries nicknames, such as shortening `numpy` to `np` as this saves us time typing out the full name when we use the library's functions.

```python
# numpy is a very helpful library for maths and numerical tasks
import numpy as np

# matplotlib is a very helpful library for plotting things
import matplotlib.pyplot as plt
```

To use a function from a particular library:
* Write the name or nickname of the library, eg. `np` or `plt`
* Use a full stop `.`
* Write the function we want to use from it

For example,

```python
# "linspace" function from numpy makes a list of numbers with definition: (start, end, how many numbers in list)
x = np.linspace(0, 15, 100)
```

Use the cell below to print out the contents of this list.

```python
#################################
### REPLACE ME WITH YOUR CODE ###
#################################
```

```python
#@title Solution:
print(x)
```

Plotting graphs is made super easy by using a library.

If we want to plot a sin(x) curve, the functions inside the libraries we imported make this very easy!

Remember we have nicknamed our libraries `np` and `plt`, and to call a function from inside a library we need the nickname, a `.` , and then the name of the function.

```python
# we can make a variable for sin(x) and call it y
y = np.sin(x)

# plot the function
plt.plot(x, y)

# add axes and title to the plot
plt.title("A sine curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```
